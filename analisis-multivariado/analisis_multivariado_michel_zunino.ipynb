```python
#Importamos las librerias
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.head()
```

```python


#Correlaciones
plt.figure(dpi = 120,figsize= (5,4))
mask = np.triu(np.ones_like(df.corr(),dtype = bool))
sns.heatmap(df.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()

```

En el grfico de calor podemos ver que hay una estrecha relacin entre el indice de masa corporal (BMI) y la edad.
Y la variable de edad es la ms determinante con respecto a todas las dems.

```python
#Crear el Pairplot pero ahora con el atributo hue = 'Outcome'. Las leyendas debern decir: 'Non Diabetic','Diabetic'
plt.figure(dpi = 120)
sns.pairplot(df,hue = 'stroke',palette = 'plasma')
plt.legend(['Stroke','No Stroke'])
plt.show()
```

**Insight**: En este grafico tambin se puede ver la misma relacin que existe entre los BMIs y la edad, as tambin entre los niveles de glucosa y BMI

```python
g = sns.jointplot(data=df, x="age", y="bmi")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```
